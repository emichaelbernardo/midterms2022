```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('NMF_df.csv')
df
```

```python
df.columns
```

```python
newdf = df[['User',
            'Content',
            'Favs', 
            'RTs',
            'Party',
            'Topic_name']].copy()
```

```python
def scale_RTs(df):
    conditions = [
        df['RTs'] == 0,
        (df['RTs'] > 0) & (df['RTs'] < 50),
        (df['RTs'] > 50) & (df['RTs'] < 500),    
        (df['RTs'] > 500) & (df['RTs'] < 5000),
        (df['RTs'] > 5000),
        ]
    values = [0,1,2,3,4]
    df['RT_scaled'] = np.select(conditions, values, default=0)    

def scale_Favs(df):
    conditions = [
        df['Favs'] == 0,
        (df['Favs'] > 0) & (df['RTs'] < 10),
        (df['Favs'] > 10) & (df['RTs'] < 100),    
        (df['Favs'] > 100) & (df['RTs'] < 1000),
        (df['Favs'] > 1000),
        ]
    values = [0,1,2,3,4]
    df['Favs_scaled'] = np.select(conditions, values, default=0)    
    
def get_reach(df):
    scale_RTs(df)
    scale_Favs(df)
    df['Reach'] = df.RT_scaled + df.Favs_scaled    
    
```

```python
get_reach(newdf)
```

```python
newdf.columns
```

```python
newdf = newdf[['User',
            'Content',
            'Favs', 
            'RTs',
            'Reach',
            'Party',
            'Topic_name']].copy()

newdf = newdf[newdf.Reach>3].count()
```

```python


newdf.to_pickle('newdf.pickle')
newdf.to_csv('newdf.csv', encoding='utf-8', index=False)
```
